```python
headlines=[]
for i in range(0,len(data)):
    X=" ".join(str(x) for x in data.iloc[i,:])
    headlines.append(X)
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
```

```python
TFIDF=TfidfVectorizer(ngram_range=(2,2))
```

```python
traindataset=countvector.fit_transform(headlines)
```

```python
traindataset=TFIDF.fit_transform(headlines)
```

```python
##GridSearchCV and RandomisedSearchCV
estimators=[int(i) for i in np.linspace(200,800,6)]
#max_depth=[int(i) for i in np.linspace(5,9,5)]
```

```python
randf=RandomForestClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
mod=RandomizedSearchCV(randf,verbose=1,cv=5,random_state=20,n_jobs=-1,param_distributions={'n_estimators':estimators})
```

```python
mod1=GridSearchCV(randf,verbose=1,cv=5,n_jobs=-1,param_grid={})
```

```python
mod.fit(traindataset,train['Label'])
```

```python
data1=data1.reset_index(drop=True)
```

```python
data1.head(2)
```

```python
test_headlines=[]
for i in range(0,len(data1)):
    X=" ".join(str(x) for x in data1.iloc[i,:])
    test_headlines.append(X)
```

```python
testdataset=TFIDF.transform(test_headlines)
```

```python
predictions=mod.predict(testdataset)
```

***Performance of the model***


```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(predictions,test['Label'])
```
